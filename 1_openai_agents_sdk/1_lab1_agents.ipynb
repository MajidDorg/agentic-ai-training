## The three pieces of an agent

An agent needs a **name**, **instructions** (its system prompt), and a **model**. Then we hand it a **user prompt** and `run` it.

```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace

# Loads OPENAI_API_KEY from your .env file
load_dotenv(override=True)
```

```python
agent = Agent(
    name="Assistant",
    instructions="You are a concise, helpful assistant.",  # the system prompt
    model="gpt-5.4-mini",
)

# 'Tell me a fun fact about agents' is the user prompt.
result = await Runner.run(agent, "Tell me one fun fact about AI agents.")
print(result.final_output)
```

That's a working agent. Notice the line started with `await` — that's the **async** part. Let's understand it.

## Async, in one idea

Talking to a model takes time — a second or more. **Async** means your program doesn't sit frozen while it waits; it can do other things, like talk to *several* agents at once.

`Runner.run(...)` is an **async** function, which is why we put `await` in front of it. In a notebook you can use `await` directly. (In a normal `.py` script you wrap it in `asyncio.run(...)` — you'll see that later.)

Here's the payoff: running three agents **in parallel** with `asyncio.gather`.

```python
import asyncio

optimist = Agent(name="Optimist", instructions="Answer in one upbeat sentence.", model="gpt-5.4-mini")
skeptic  = Agent(name="Skeptic",  instructions="Answer in one cautious sentence.", model="gpt-5.4-mini")
realist  = Agent(name="Realist",  instructions="Answer in one balanced sentence.", model="gpt-5.4-mini")

question = "Will AI agents change how people work?"

# 'trace' groups these runs together so you can inspect them as one unit.
with trace("Three views in parallel"):
    results = await asyncio.gather(
        Runner.run(optimist, question),
        Runner.run(skeptic, question),
        Runner.run(realist, question),
    )

for r in results:
    print(r.final_output, "\n")
```

All three ran at the same time instead of one-after-another. That's why agentic systems lean on async — they often coordinate several model calls at once.

## See what happened: traces

Open **https://platform.openai.com/traces** and find **"Three views in parallel"**. You'll see each agent run inside it. Get into the habit of checking traces — it's how you understand what your agents actually did.

## Recap
- An agent = **name + instructions + model**, run with `await Runner.run(agent, prompt)`.
- `Runner.run` is **async**; `asyncio.gather` runs many agents in parallel.
- **Traces** let you inspect every run.
